```python
import os

def rename_multiple_files(path,obj):
    i=0
    for filename in os.listdir(path):
        try:
            f,extension = os.path.splitext(path+filename)
            src=path+filename
            dst=path+obj+str(i)+extension
            os.rename(src,dst)
            i+=1
            print('Rename successful.')
        except:
            i+=1
```

```python
for i in os.listdir('Phill-Images(class_name)/'):
    if i == '.DS_Store': continue
    else: rename_multiple_files(f'Phill-Images(class_name)/{i}/', i)
```
